```python
df.isnull().sum()                 #Check missing values, so that it can be replaced, don't just replace it with median, try finding the median in each category and then try to replace with that
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')                 #convert all headings into snakecase
```

```python
df.columns
```

```python
df = df.rename(columns={'purchase_amount_(usd)' : 'purchase_amount'})        #make changes in column purchase_amount_(usd)
```

```python
df.columns
```

```python
#create column age_group
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)             #qcut splits ages in 4 equal-sized group
```

```python
df[['age','age_group']].head(10)
```

```python
#create column purchase freaquency days
frequency_mapping = {
 'Fortnightly':14 ,
 'Weekly':7,
 'Monthly':30,
 'Quarterly':90,
 'Bi-Weekly':14,
 'Annually':365, 
 'Every 3 Months':90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10) 
```

```python
#check for the duplicate columns or columns having same values so we needto drop one of the column
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
df = df.drop('promo_code_used',axis=1)
```
